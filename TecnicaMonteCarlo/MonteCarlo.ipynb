1.Utilizar tecnicas de monte-carlo para encontar una aproximación de la solución de una integral

```python
#IMPORTACION DE LIBRERIAS
import random
import matplotlib.pyplot as plt 
import numpy as np # importando numpy
import pandas as pd # importando pandas
from scipy import stats
import math

print('LIBRERIAS IMPORTADAS')
```

```python
#CREACION DE VARIABLES PARA EL ALMACENAMIENTO Y GRAFICO DE LA INTEGRAL
n = 800
aleatorios=[]
fx=[]
rnd=[]
summary=[]
x=[]
y=[]
random.seed(123222)
```

```python
for i in range(0, n):
  value=random.uniform(0.0, 1.0)
  aleatorios.append(value)
  fx.append(1-value**2)
  rnd.append(random.uniform(0.0, 1.0))

for i in range(0, n):
  selected=1 if rnd[i]<=fx[i] else 0
  if(selected==1):
    x.append(aleatorios[i])
    y.append(rnd[i])
  summary.append(selected)
  
#CALCULO DEL AREA ESTIMADA
areaEstimada=(summary.count(1)/n)
print('Area Estimada= ', areaEstimada)
```

```python
#GRAFICA CON TODOS LOS VALORES ALEATORIOS Y LA ECUACION
plt.plot(aleatorios,fx, 'b.')
plt.plot(aleatorios,rnd,'o')
```

```python
#GRAFICA CON LOS VALORES QUE SATISFACEN LA INTEGRAL
plt.plot(aleatorios,fx, 'r.')
plt.plot(x,y,'o')
```

```python
aleatorios=[]
fx=[]
gx=[]
rnd=[]
summary=[]
x=[]
y=[]

for i in range(0, n):
  value=random.uniform(0.0, 1.0)
  aleatorios.append(value)
  fx.append(value**2)
  gx.append(math.sqrt(value))
  rnd.append(random.uniform(0.0, 1.0))

for i in range(0, n):
  selected=1 if rnd[i]>=fx[i] and rnd[i] <= gx[i] else 0
  if(selected==1):
    x.append(aleatorios[i])
    y.append(rnd[i])
  summary.append(selected)
#CALCULO DEL AREA ESTIMADA DE LA INTEGRAL
areaEstimada=(summary.count(1)/n)
print('Area Estimada= ', areaEstimada)
```

```python
#GRAFICA CON TODOS LOS VALORES ALEATORIOS Y LAS ECUACIONES
plt.plot(aleatorios,fx, 'b.')
plt.plot(aleatorios,gx, 'r.')
plt.plot(aleatorios,rnd,'o')
```

```python
#GRAFICA CON LOS VALORES QUE SATISFACEN LA INTEGRAL
plt.plot(aleatorios,fx, 'b.')
plt.plot(aleatorios,gx, 'r.')
plt.plot(x,y,'o')
```
